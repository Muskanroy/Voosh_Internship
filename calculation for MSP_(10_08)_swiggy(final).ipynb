```python
import os
import pandas as pd
cwd = os.path.abspath(r'C:/Users/muska/OneDrive/Desktop/Voosh_intern/invoice_download') 
files = os.listdir(cwd)

import pandas as pd
import glob
mydict={}
for i in range(len(files)):
    b=files[i]
#     path of the folder
    path = r'C:/Users/muska/OneDrive/Desktop/Voosh_intern/invoice_download/'+'{}'.format(b) # use your path

# display the list of excel files
    all_files = glob.glob(path + "/*.xlsx")
    
    # Intializing the empty data frame
    finaldf=pd.DataFrame()

    # To iterate excel file one by one inside the folder
    for filename in all_files:

    # Combining multiple excel sheets into one dataframe
        dct =pd.concat(pd.read_excel(filename,sheet_name=[1],names=['s.no','section','Delivered Orders','Cancelled Orders','Total']),ignore_index=True)
        finaldf=finaldf.append(dct,ignore_index=True)
    # convert datatype of column "Total" of DataFrame to numeric
    finaldf["Total"] = pd.to_numeric(finaldf["Total"],errors='coerce')
#     mydict={}
    list1=[]
    Total_Customer_Payable=finaldf.loc[finaldf['s.no'] == 'A  ', 'Total'].sum()
    list1.append(Total_Customer_Payable)
    Total_Swiggy_Service_Fees=finaldf.loc[finaldf['s.no'] == 'B ', 'Total'].sum()
    list1.append(Total_Swiggy_Service_Fees)
    Order_Level_Adjustments=finaldf.loc[finaldf['s.no'] == 'C', 'Total'].sum()
    list1.append(Order_Level_Adjustments)
    Merchant_Level_Adjustments=finaldf.loc[finaldf['s.no'] == 'D', 'Total'].sum()
    list1.append(Merchant_Level_Adjustments)
    try:
        High_Priority=finaldf.loc[finaldf['section'] == 'High Priority', 'Total'].sum()
        list1.append(High_Priority)
    except:
        pass
    try:
        Homepage_Banner=finaldf.loc[finaldf['section'] == 'Homepage Banner', 'Total'].sum()
        list1.append(Homepage_Banner)
    except:
        pass
    try:
        Onboarding_Fees=finaldf.loc[finaldf['section'] == 'Onboarding Fees', 'Total'].sum()
        list1.append(Onboarding_Fees)
    except:
        pass
    try:
        Refund=finaldf.loc[finaldf['section'] == 'Refund', 'Total'].sum()
        list1.append(Refund)
    except:
        pass
    try:
        Long_Distance_Pack_Fee=finaldf.loc[finaldf['section'] == 'Long Distance Pack Fee', 'Total'].sum()
        list1.append(Long_Distance_Pack_Fee)
    except:
        pass
    try:
        Excess_payout=finaldf.loc[finaldf['section'] == 'Excess payout', 'Total'].sum()
        list1.append(Excess_payout)
    except:
        pass
    Net_Payout_Before_TCS=finaldf.loc[finaldf['s.no'] == 'E ', 'Total'].sum()
    list1.append(Net_Payout_Before_TCS)
    TCS=finaldf.loc[finaldf['s.no'] == 'F ', 'Total'].sum()
    list1.append(TCS)
    TDS=finaldf.loc[finaldf['s.no'] == 'G', 'Total'].sum()
    list1.append(TDS)
    t=list(finaldf.loc[finaldf['s.no'] == 'H ']['Total'])
    list2=[]
    list3=[]
    for i in t:

        if i<0:
            list2.append(i)
        else:
            list3.append(i)
    if len(list3)==0:
        Net_Payout=list2[-1]
    else:
        Net_Payout=sum(list3)
    # Net_Payout=finaldf.loc[finaldf['s.no'] == 'H ', 'Total'].sum()
    list1.append(Net_Payout)
    list1
    #for i in range(len(files)):
        #b=files[i]
    mydict[b]=list1
print(mydict)
    
```
