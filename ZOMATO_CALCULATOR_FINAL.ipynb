```python
for i in voosh:
    temp = data[data['voosh_id']==i]
    d = {}
    d.setdefault('voosh_id',int(i))
    for j in name_list[4:]:
        if j!='commission_pct':
            ans = temp[j].sum()
            d.setdefault(j,ans)
        else:
            ans = temp[j].sum()/len(temp)
            d.setdefault(j,ans)
    new_df = new_df.append(d,ignore_index=True)
```
